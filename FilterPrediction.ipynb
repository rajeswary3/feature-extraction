```python
def chkIfAlreadyExistExactMatch(word,strArr):

    
    strW = strArr.split("|")
   
    for h in range (len(strW)):
        strWW = strW[h]
        print("strWW ",strWW)
        print ( "word ",word)
        wExist= "false"
        if ( word == strWW):
            print(word ,"exist", strWW)
            wExist ="true"
            break
     
    return wExist;
```

```python
# read freq nouns
freqNounList=[]
with open("NounSetsValuefreq.txt", "r") as file:
    for line in file:
        s = line.rstrip()  
        s.strip('"')
        if (len(s)>0):
            s= s.split("|")
            nT =s[0]
            t= s[1]
            tInt = int(t)
            if (tInt>=0):
                freqNounList.append(nT)
file.close()

def chkFreqNoun(word) :
    exist = ""
    op =""
    for a in range ( len (freqNounList)):
        op = freqNounList[a]
        if ( op== word):
            exist ="true"
            break
        else:
            exist ="false"   
    return "true";
    
```
