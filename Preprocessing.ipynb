```python
## to filter out special chars except for comma ''' "()" '/""'

def text_to_word_sequence(text,filters='-!"#$%*+.:;<=>?@[\\]^_`{|}~\t\n)(',lower=True, split=" "):
    if lower:
        text = text.lower()
    text = text.translate(str.maketrans(filters, split * len(filters)))
    #seq = text.split(split)
    #return [i for i in seq if i] 
    return text;
```

```python
reviews=[]
with open("Nokia6610Reviews_Explicit_246.txt", "r") as file:
    for line in file:
        #f.read()
        #print(line)
        s = line.rstrip()  
        s.strip('"')
        #print(s)
        if (len(s)>2):
            s= s.split("|")
            #print(s[0])
            t= s[1]
            t=t.lstrip('\"')
            #print(t)
            #print(s[1])
            reviews.append([s[0],t])
        
file.close()
```

```python
import AddFunctionalities as adjF
cleanedRS =[]

for tK in range (len (reviews)):
   
    reviewId  =reviews[tK][0]
    reviewSentence = reviews[tK][1]
    reviewSentence = reviewSentence.lower()
    reviewSentence = re.sub(r"\d+"," " , reviewSentence)
    txt =""
    tempS= reviewSentence.split(" ")
    for d in range (len(tempS)):
        wd  = tempS[d]
        wd = convertShortExp(wd)
        txt = txt+" "+wd
    reviewSentence = txt
    """To remove punctuations """
    reviewSentence=text_to_word_sequence(reviewSentence)
    """To remove whitespacec"""
    reviewSentence = reviewSentence.strip()
    '''Token and stop word removals'''
    filtered_sentence = ""
    word_tokens = reviewSentence.split()
    for w in word_tokens: 
        if w not in my_stopwords: 
            filtered_sentence= filtered_sentence + " "+w 
    reviewSentence=filtered_sentence
    reviewSentence =spell (reviewSentence)
    """To replace / to or """
    txt =""
    tempS= reviewSentence.split(" ")
    for d in range (len(tempS)):
        wd  = tempS[d]
        wd1 = wd.replace("/"," or ")
        txt = txt+" "+wd1
    reviewSentence = txt
    st = reviewId+"|"+reviewSentence
    cleanedRS.append(st)
data=cleanedRS

df = pd.DataFrame(data,columns=['Review'])
np.savetxt(r'NokiaCleanReviews_246.txt', df.values, fmt='%s',delimiter ='|')
```
